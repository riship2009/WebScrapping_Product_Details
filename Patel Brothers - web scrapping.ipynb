```python
from urllib.request import urlopen as ureq
from bs4 import BeautifulSoup as soup

pb_url = 'https://patelbrothersusa.com/groceries/new-category/?sort=featured&limit=100&mode=6'
pb_url
```

```python

a = ureq(pb_url) 
pb_page = a.read()  
a.close()  

pb_page
```

```python

pb_html = soup(pb_page, "html.parser")
pb_html
```

```python
b = pb_html.findAll("div",{"class":"card-body"})
len(b)
```
